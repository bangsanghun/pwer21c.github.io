## 거북이 만들기입니다. 몰라도 일단 그대로 써봅시다. 다음에 이것이 무엇인지를 설명할것입니다.

```python
import turtle
t=turtle.Turtle()
t.shape('turtle')
```

## 거북이 앞으로 100걸음 가기

```python
t.forward(100)
```

## 거북이가 왼쪽으로 90도 몸을 틀어서 앞으로 100걸음 갑니다.

```python
t.left(90)
t.forward(100)
```

## 거북이를 다시 오른쪽으로 몸을 틀어서 앞으로 100걸음 갑니다.

```python
t.right(90)
t.forward(100)
```

## 거북이를 다시 오른쪽으로 몸을 틀어서 앞으로 100걸음 갑니다.

```python
t.right(90)
t.forward(100)
```

## 거북이가 왼쪽으로 90도 몸을 틀어서 앞으로 100걸음 갑니다.

```python
t.left(90)
t.forward(100)
```

## 그림을 지우고 다시 시작하는 명령어입니다.

```python
t.reset()
```

## 원을 그려요

```python
t.circle(100)
```

## 세번째 그림

```python
t.reset() ## 그림을 지웁니다.
t.forward(100)  ## 앞으로 100걸음 간다.
t.circle(100) ## 한바퀴 돌아요
t.forward(100) ## 앞으로 100걸음 갑니다.
```

거북이를 움직여 봅시다
forward : 앞으로 
backward: 뒤로
left: 왼쪽으로
right: 오른쪽으로
    

## 재미있는 거북이

```python
## 빙글빙글 세번 도는 거북이.
t.reset()
t.circle(10)
t.circle(20)
t.circle(30)
```
